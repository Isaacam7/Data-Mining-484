```python
import pandas as pd
import pickle as pk
mergedfs = pk.load(open('mergedfs.pkl', 'rb'))
```

```python
#for math and rla proficiencies reformat the names
for i in range(10):
    if i == 10:
        continue
    if i == 0:
        mergedfs[i]["ALL_MTH00pctprof"] = mergedfs[i][f"ALL_MTH00pctprof_0{9 + i}{9 + (i+1)}"]
        mergedfs[i]["ALL_RLA00pctprof"] = mergedfs[i][f"ALL_RLA00pctprof_0{9 + i}{9 + (i+1)}"]
        continue
    mergedfs[i]["ALL_MTH00pctprof"] = mergedfs[i][f"ALL_MTH00pctprof_{9 + i}{9 + (i+1)}"]
    mergedfs[i]["ALL_RLA00pctprof"] = mergedfs[i][f"ALL_RLA00pctprof_{9 + i}{9 + (i+1)}"]
    
```

```python
#add important vars that need to be kept for the dataset
Vars = pk.load(open('vars.pkl', 'rb'))
Vars.extend(['STATE','NCESID','ALL_MTH00pctprof','ALL_RLA00pctprof','YRDATA','ENROLL','Fall_membership'])
Vars.extend(['TOTALREV', 'TFEDREV', 'FEDRCOMP', 'FEDRSPEC', 'FEDRNUTR', 'FEDROTHR',
       'TSTREV', 'STRFORM', 'STRSPEC', 'STRTRANS', 'STROTHR', 'TLOCREV',
       'LOCRTAX', 'LOCRPROP', 'LOCRPAR', 'LOCRCICO', 'LOCROSCH', 'LOCRCHAR',
       'LOCROTHR', 'TOTALEXP', 'TCURSPND', 'TSALWAGE', 'TEMPBENE', 'TCURINST',
       'TCURISAL', 'TCURIBEN', 'TCURSSVC', 'TCURSPUP', 'TCURSSTA', 'TCURSGEN',
       'TCURSSCH', 'TCURSOTH', 'TCURONON', 'TCAPOUT', 'TPAYOTH', 'TINTRST',
       'DEBTOUT', 'LONGISSU', 'LONGRET', 'PCTTOTAL', 'PCTFTOT', 'PCTFCOMP',
       'PCTSTOT', 'PCTSFORM', 'PCTLTOT', 'PCTLTAXP', 'PCTLOTHG', 'PCTLCHAR',
       'PPCSTOT', 'PPSALWG', 'PPEMPBEN', 'PPITOTAL', 'PPISALWG', 'PPIEMBEN',
       'PPSTOTAL', 'PPSPUPIL', 'PPSSTAFF', 'PPSGENAD', 'PPSSCHAD','TCUROTH','NONELSEC'])

#remove the vars not to be used
removelist = ['Fed_rev_state_Title_II_Part_A', 'Fed_rev_state_Title_IV_Part_A_1', 'Fed_rev_state_Title_IV_Part_B', 'Fed_rev_state_Title_V_Part_B_2', 'Direct_fed_rev_Title_V_Part_B_1',
              'CARES_Act_Funds_Rev_ESSER_Fund', 'CARES_Act_Funds_Rev_GEER_Fund', 'CARES_Act_Funds_Rev_ESF_RWP_Discretionary_Grant', 'CARES_Act_Funds_Rev_ESF_REM_Discretionary_Grant',
              'CARES_Act_Funds_Rev_Project_SERV', 'TSTREV', 'TOTALREV','TFEDREV', 'TOTALEXP','TLOCREV',
              'CARES_Act_Funds_Rev_CRF', 'CARES_Act_Funds_Exp_Current_spending', 'CARES_Act_Funds_Exp_Instructional',
              'CARES_Act_Funds_Exp_Support_services', 'CARES_Act_Funds_Exp_Capital_outlay', 'CARES_Act_Funds_Exp_Tech_supplies_purchased_svcs', 'CARES_Act_Funds_Exp_Tech_equipment']


Vars = list(set(Vars) - set(removelist))
```

```python
#create missing sum sections for dataset
TCURSSVC = {
    "E17": "Current_exp-Pupil_support",
    "E07": "Current_exp-Instruction_staff",
    "E08": "Current_exp-General_admin",
    "E09": "Current_exp-School_admin",
    "V40":"Curr_opex_O&M_plant",
    "V45":"Curr_opex_Student_trans",
    "V90":"Curr_opex_Bus_cent_other",
    "V85":"Curr_opex_Nonspec_supp",
    "J17":"State_pay_Pupil_supp_ben",
    "J07":"State_pay_Instr_staff_ben",
    "J08":"State_pay_Gen_admin_ben",
    "J09":"State_pay_Sch_admin_ben",
    "J40":"State_pay_O&M_plant_ben",
    "J45":"State_pay_Student_trans_ben",
    "J90":"State_pay_Bus_cent_other_ben",
    "J11":"Own_ret_sys_trans_Supp",
    "J96":"State_pay_Supp_nonben"
}

TCAPOUT = {
    "F12": "Construction",
    "J98":"State_pmt_on_behalf_Nonel_sec_prog",
    "G15": "Purchase_of_land_and_existing_structures",
    "K09": "Instructional_equipment",
    "K10": "Other_equipment",
    "K11": "Nonspecified_equipment",
    "J99": "State_payments_on_behalf_-_Capital_outlay",
}

TCURINST = {
    "E13": "Current_exp-Instruction",
    "J13": "State_pmts-Instruction_benefits",
    "J12": "Own_retirement_sys-Instruction",
    "J14": "State_pmts-Instruction_nonben",
    "V91": "Exhibit-Payments_private_schls",
    "V92": "Exhibit-Payments_charter_schls",
}

TCUROTH = {
    "E11":"Curr_opex_Food_services",
    "V60":"Curr_opex_Enterprise_ops",
    "V65":"Curr_opex_Oth_Elem_Sec_prog",
    "J10":"State_pay_Other_ben",
    "J97":"State_pay_Noninstr_nonben",
}

NONELSEC = {
    "V70":"Curr_opex_Comm_services",
    "V75":"Curr_opex_Adult_educ",
    "V80":"Curr_opex_Oth",
    "F12": "Construction",
    "J98":"State_pmt_on_behalf_Nonel_sec_prog",
}



#add missing sum columns
for i, df in enumerate(mergedfs):
    if i == 10:
        continue
    #create the missing sum columns
    df.loc[:, 'TCURSSVC'] = df[list(TCURSSVC.values())].sum(axis=1)
    df.loc[:, 'TCAPOUT']  = df[list(TCAPOUT.values())].sum(axis=1)
    df.loc[:, 'TCURINST'] = df[list(TCURINST.values())].sum(axis=1)
    df.loc[:, 'TCUROTH'] = df[list(TCUROTH.values())].sum(axis=1)
    df.loc[:, 'NONELSEC']  = df[list(NONELSEC.values())].sum(axis=1)
```

```python
#create other category columns
Total_sal_and_wages = {
    "Z33": "Total_sal_and_wage_Instruction",
    "V11": "Total_sal_and_wage_Pupil_support",
    "V13": "Total_sal_and_wage_Instructional_staff_support",
    "V15": "Total_sal_and_wage_General_administration",
    "V17": "Total_sal_and_wage_School_administration",
    "V21": "Total_sal_and_wage_Operation_and_maintenance_of_plant",
    "V23": "Total_sal_and_wage_Student_transportation",
    "V37": "Total_sal_and_wage_Business/central/other_support_services",
    "V29": "Total_sal_and_wage_Food_services",
}

Total_empl_benf_payms = {
    "V10": "Total_empl_benf_paymInstruction",
    "V12": "Total_empl_benf_paym_Pupil_support",
    "V14": "Total_empl_benf_paym_Instructional_staff",
    "V16": "Total_empl_benf_paym_General_administration",
    "V18": "Total_empl_benf_paym_School_administration",
    "V22": "Total_empl_benf_paym_Operation_and_maintenance_of_plant",
    "V24": "Total_empl_benf_paym_Student_transportation",
    "V38": "Total_empl_benf_paym_Business/central/other_support_services",
    "V30": "Total_empl_benf_paym_Food_services",
    "V32": "Total_empl_benf_paym_Enterprise_operations",
}
```

```python
import copy
#create new empty list of dataframes
scoodfs =  [pd.DataFrame() for _ in range(10)]

"""
for every dataframe in the list, calculate the proportion data for the sum columns (so for every spending source that is in a sub-category find the proportion of
sub category that the spending source is)
"""
for i in range(10):
    merged_copy = copy.deepcopy(mergedfs[i])
    scoodfs[i][['NCESID','YRDATA','STATE','Fall_membership','ALL_MTH00pctprof','ALL_RLA00pctprof']] = merged_copy[['NCESID','YRDATA','STATE','Fall_membership','ALL_MTH00pctprof','ALL_RLA00pctprof']]
    for val in list(TCURSSVC.values()):
        scoodfs[i].loc[:,f'prop_TCURSSVC_{val}'] = merged_copy.loc[:,val] / merged_copy.loc[:,'TCURSSVC']
    for val in list(TCAPOUT.values()):
        scoodfs[i].loc[:,f'prop_TCAPOUT_{val}'] = merged_copy.loc[:,val] / merged_copy.loc[:,'TCAPOUT']
    for val in list(TCURINST.values()):
        scoodfs[i].loc[:,f'prop_TCURINST_{val}'] = merged_copy.loc[:,val] / merged_copy.loc[:,'TCURINST']
    for val in list(TCUROTH.values()):
        scoodfs[i].loc[:,f'prop_TCUROTH_{val}'] = merged_copy.loc[:,val] / merged_copy.loc[:,'TCUROTH']
    for val in list(NONELSEC.values()):
        scoodfs[i].loc[:,f'prop_NONELSEC_{val}'] = merged_copy.loc[:,val] / merged_copy.loc[:,'NONELSEC']
    for val in list(Total_sal_and_wages.values()):
        valname = val.split('Total_sal_and_wage_')[-1]
        scoodfs[i].loc[:,f'prop_sal_{valname}'] = merged_copy.loc[:,val] / merged_copy.loc[:,'Total_sal_and_wages']
    for val in list(Total_empl_benf_payms.values()):
        valname = val.split('Total_empl_benf_')[-1]
        scoodfs[i].loc[:,f'prop_benf_{valname}'] = merged_copy.loc[:,val] / merged_copy.loc[:,'Total_empl_benf_payms']
    
    #here find the sub-category spending per pupil (student)
    scoodfs[i].loc[:,'TCURSSVC_spending_per_student'] = merged_copy.loc[:,'TCURSSVC'] / merged_copy.loc[:,'Fall_membership']
    scoodfs[i].loc[:,'TCAPOUT_spending_per_student'] = merged_copy.loc[:,'TCAPOUT'] / merged_copy.loc[:,'Fall_membership']
    scoodfs[i].loc[:,'TCURINST_spending_per_student'] = merged_copy.loc[:,'TCURINST'] / merged_copy.loc[:,'Fall_membership']
    scoodfs[i].loc[:,'TCUROTH_spending_per_student'] = merged_copy.loc[:,'TCUROTH'] / merged_copy.loc[:,'Fall_membership']
    scoodfs[i].loc[:,'NONELSEC_spending_per_student'] = merged_copy.loc[:,'NONELSEC'] / merged_copy.loc[:,'Fall_membership']
    scoodfs[i].loc[:,'Total_sal_and_wages_spending_per_student'] = merged_copy.loc[:,'Total_sal_and_wages'] / merged_copy.loc[:,'Fall_membership']
    scoodfs[i].loc[:,'Total_empl_benf_payms_spending_per_student'] = merged_copy.loc[:,'Total_empl_benf_payms'] / merged_copy.loc[:,'Fall_membership']
    
    #only keep school systems that have more than 200 people
    enroll_cutoff = 200
    mask = scoodfs[i]['Fall_membership'] >= enroll_cutoff
    scoodfs[i] = scoodfs[i][mask]  

    
```
